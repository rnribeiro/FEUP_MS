```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Manually specify the CSV files and their labels
csv_files = [
    'csvs/scenario_4_dqn_DQN_16_confort.csv',  # Replace with the actual path to your first file
    'csvs/scenario_4_dqn_DQN_1.csv',  # Replace with the actual path to your second file
    'csvs/scenario_4_ppo_PPO_1_confort.csv',  # Replace with the actual path to your third file
    'csvs/scenario_4_ppo_PPO_1.csv'   # Replace with the actual path to your fourth file
]

labels = [
    'DQN Comfort',  # Label for the first file
    'DQN Simple',  # Label for the second file
    'PPO Comfort',  # Label for the third file
    'PPO Simple'   # Label for the fourth file
]

# Step 2: Create a figure and axis for the plot
plt.figure(figsize=(12, 8))  # Adjust figure size
ax = plt.gca()

# Step 3: Define colors and line styles for differentiation
colors = ['b', 'g', 'r', 'c']  # Blue, Green, Red, Cyan
line_styles = ['-', '--', '-.', ':']  # Solid, Dashed, Dash-dot, Dotted

# Step 4: Loop through each CSV file and plot the data
for i, (file, label) in enumerate(zip(csv_files, labels)):
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Plot the data
    ax.plot(
        df['Step'], 
        df['Value'], 
        label=label,  # Use the custom label
        color=colors[i % len(colors)],  # Cycle through colors
        linestyle=line_styles[0],  # Cycle through line styles
        linewidth=2  # Adjust line thickness
    )

# Step 5: Add labels, title, and legend
ax.set_xlabel('Steps', fontsize=14)
ax.set_ylabel('Reward', fontsize=14)
ax.legend(loc='upper right', fontsize=12)  # Add legend

# Step 6: Customize the grid
ax.grid(True, linestyle='--', alpha=0.6)  # Add a dashed grid with transparency

# Step 7: Save the plot to a file (optional)
plt.savefig('reward_vs_steps.png', dpi=300, bbox_inches='tight')  # Save as PNG

# Step 8: Show the plot
plt.show()
```
