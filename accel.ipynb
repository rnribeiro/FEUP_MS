```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the acceleration data from the file
data = pd.read_csv('acceleration.txt', header=None, names=['acceleration'])

print(max(data['acceleration']))

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.hist(data['acceleration'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Acceleration Distribution')
plt.xlabel('Acceleration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# plot line graph
plt.figure(figsize=(10, 6))
plt.plot(data['acceleration'])
plt.title('Acceleration Line Graph')
plt.xlabel('Time')
plt.ylabel('Acceleration')
plt.grid(True)
plt.show()
```

```python
# plot the the squared acceleration values
plt.figure(figsize=(10, 6))
plt.plot(data['acceleration']**2)
plt.title('Acceleration Squared Line Graph')
plt.xlabel('Time')
plt.ylabel('Acceleration Squared')
plt.grid(True)
plt.show()

```

```python
# Calculate the derivative of the acceleration data
data['derivative'] = data['acceleration'].diff()

# Plot the derivative graph
plt.figure(figsize=(10, 6))
plt.plot(data['derivative'])
plt.title('Derivative of Acceleration')
plt.xlabel('Time')
plt.ylabel('Derivative of Acceleration')
plt.grid(True)
plt.show()
```
